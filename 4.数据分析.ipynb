```python
#数据变形
import pandas as pd
data = {
'ID': [1, 2],
'name':['alice','bob'],
'Math':[90,85],
'English': [88, 92],
'Science': [95, 89]
}
# 2. 创建DataFrame
df = pd.DataFrame(data)

# 3. 宽表转长表（正确用法）
df_long = pd.melt(
    df,
    id_vars=['ID', 'name'],   # 保留的列（注意这里列名是'name'，和上面data里的键一致）
    var_name='科目',           # 原来的列名（Math/English/Science）会变成这一列的值
    value_name='成绩'          # 原来的列值会变成这一列的值
)
df_long.sort_values('name')
# 查看结果
```

```python
#分组聚合
#df.group(‘分组的字段’)['聚合的字段']
df = pd.read_csv('data/employees.csv')
print(df['department_id'].isna().sum())#查找是否有缺失值
df= df.dropna(subset=['department_id'])#删除这一列有缺失值的
print(df['department_id'].isna().sum())
df['department_id'] = df['department_id'].astype('int64')
#计算不同部门的平均薪资
df.groupby('department_id').groups#查看分组情况
df.groupby('department_id').get_group(10)#查看具体的某个分组数据
df.groupby('department_id')[['salary']].mean()
```

```python
#企鹅数据分析
#1.导入必要的库
import pandas as pd
import numpy as np
#2.导入数据 喙
df = pd.read_csv('data/penguins.csv')
df.head(5)
df.info()
```

```python
#1.导入库
import pandas as pd
#2.导入数据
df = pd.read_csv('data/sleep.csv')
df.head()
df.info()
df.describe()
```

```python
#3.数据清洗
df.isna().sum()
# 方法1：先判断列是否存在，再删除
if 'sleep_disorder' in df.columns:
    df.drop(columns='sleep_disorder', inplace=True)

# 方法2：用 errors='ignore' 参数，不存在也不报错
df.drop(columns='sleep_disorder', inplace=True, errors='ignore')
df.isna().sum()
```

```python
#4.数据转换，数据特征构造
#先处理字符串，因为字符串类型的文本，模型看不懂，没法直接计算，必须转成数字 / 分类类型才能分析、建模。
df['gender'] = df['gender'].astype('category')
# df.info()#字符串已经变成分类了
df.head()
df['occupation'].value_counts()#先看能不能分类： 什么时候适合转 category？
# 满足这两个条件：
# 取值是固定选项：比如性别、职业、学历、等级、城市等，只有有限的几种值。
# 重复率高：比如 1000 行数据里，只有 5 个不同的值，重复越多，省内存的效果越明显。
df['occupation'] = df['occupation'].astype('category')
df[['high','low']]=df['blood_pressure'].str.split('/',expand=True)


#睡眠质量分箱
labels = ['差','中','优']
df['quality_level'] = pd.cut(df['sleep_quality'],bins=3,labels=labels)
age_labels=['青少年','中年','老年']
df['age_level'] = pd.cut(df['age'],bins=3,labels=age_labels)
df.head()
```

```python
#5.数据分析与统计
print(df['bmi_category'].value_counts())
```

```python
#根据不同BIMI分组，睡眠质量
df.groupby('bmi_category').agg({
    'sleep_duration':['mean','std'],
    'sleep_quality':['mean','std'],
})
```
