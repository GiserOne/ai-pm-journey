题目一：温度数据分析
某城市一周的最高气温为【29，30，29，31，32，30，29】
计算平均气温、最高气温和最低气温
找出气温超过30°的天数

```python
temps = np.array([29,30,29,31,32,30,29])
print('平均气温')
print('%.3f'%np.mean(temps))
print('最高气温')
print(np.max(temps))
print(np.min(temps))
print('气温超过三十度天数')
print(temps[temps>30])
print(len(temps[temps>30]))
arr = np.where(temps>30,temps,0)
target = 6
count = (arr > target).sum()
print(count)
print(np.where(temps>30,temps,0))
print(np.cumsum(np.where(temps>30,1,0))[-1])
```

题目 7：统计函数应用

- 找出销售额最高的月份和最低的月份。

题目 20：综合应用
某商店 5 天的销售额（万元）和成本（万元）如下：
销售额：[20, 25, 22, 30, 28]
成本：[15, 18, 16, 22, 20]

- 计算每天的利润（销售额 - 成本）。
- 计算利润的平均值和标准差。
- 找出利润最高的天数。

```python
money = np.array([20, 25, 22, 30, 28])
a = np.array([15, 18, 16, 22, 20])
print('每天的利润',money-a)
b = money-a
print('利润的平均值',np.mean(b))
print('利润的标准差',np.std(b))
print('利润最高的天数',len(b[b==np.max(b)]))
```

 什么情况下想到布尔索引
1. 筛选 / 找出满足条件的元素
大于、小于、等于、满足多个条件
arr [arr> 10] # 找出大于 10 的
arr [arr == 5] # 找出等于 5 的
arr [(arr>3) & (arr<8)] # 多个条件
2. 根据条件替换值
想把数组里满足条件的数改成别的值，比如：
大于 10 的改成 0
奇数改成 -1
低于平均值的改成平均值
arr [arr> 10] = 0 # 大于 10 的全部变 0
3. 统计满足条件的数量
有多少个数 > 5
有多少天利润 > 0
np.sum (arr> 5) # True=1, False=0, 求和就是数量
4. 获取满足条件的位置（索引）
np.where(arr == arr.max())
或者
arr == arr.max()
