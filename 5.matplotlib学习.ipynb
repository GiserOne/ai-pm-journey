```python
#绘制折线图
import  matplotlib.pyplot as plt
from matplotlib import rcParams, colors  # 字体
rcParams['font.sans-serif'] = ['SimHei']
rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
#要绘制的数据
product = ['鱼缸','龙缸','特殊岩板龙缸','特殊岩板鱼缸']
sales = [10000,20000,2000,50000]
# 给每个产品分配数字坐标（解决文本当X轴的问题）
x_pos = range(len(product))
#绘制折线图
plt.plot(x_pos,sales,
         label='产品A',marker='o',linestyle='--',color='blue')
print(sales)
#添加标题
plt.title('2025年鱼缸销售趋势',color='red',fontsize=20)
#添加坐标轴的标签
plt.xlabel('产品',fontsize = 10)
plt.ylabel('销售额（元）',fontsize = 10)
#添加图例
plt.legend(loc='upper left',fontsize = 10)
#添加网格线
plt.grid(True,alpha=0.1,color='blue',linestyle='--')
#设置刻度字体大小
plt.xticks(x_pos,product,rotation=0,fontsize=10)
#设置y轴的范围
plt.ylim(0,max(sales) * 1.2)# 自动根据数据最大值设置上限，避免点贴边
#在每个数据点上显示数值（用数字坐标x_pos,不会报错）
for i,y in enumerate(sales):#enumerate(sales) 会同时返回两个值：i：数据的索引（0,1,2,3…）。y：对应位置的销售额数值（10000, 20000, 2000, 50000）。这行代码的作用是：遍历 sales 列表里的每一个值，同时拿到它的索引和数值。
    plt.text(x_pos[i],y+1000,str(y),ha='center',va='bottom',fontsize=10)#plt.text(x, y, text) 是在图上添加文字的函数
#显示图表
plt.show()
```

```python
#绘制柱状图
import  matplotlib.pyplot as plt
from matplotlib import rcParams#字体
rcParams['font.sans-serif'] = ['SimHei']
rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
#要绘制的数据
subjects = ['语文','数学','英语','科学']
score = [85,92,78,88]
#绘制柱状图
plt.bar(subjects,score,color='orange',label='小明')
# #添加标题
plt.title('小明2027成绩分布',color='red',fontsize=20)
# #添加坐标轴的标签
plt.xlabel('科目',fontsize = 10)
plt.ylabel('分数',fontsize = 10)
# #添加图例
plt.legend(loc='upper right')
# #添加网格线
plt.grid(axis='y',alpha=0.1,color='blue',linestyle='--')
# #设置刻度字体大小
plt.xticks(rotation=0,fontsize=12)
plt.yticks(rotation=0,fontsize=12)
# #设置y轴的范围
plt.ylim(0,100)
#在每个点上显示数值
for x,y in zip(subjects,score):
    plt.text(x,y+1,str(y),ha='center',va='bottom',fontsize=10)
#默认自动优化排版的函数
plt.tight_layout()
plt.show()
```

```python
#绘制环形图
import  matplotlib.pyplot as plt
from matplotlib import rcParams#字体
rcParams['font.sans-serif'] = ['SimHei']
rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
#创建列表、设置大小
plt.figure(figsize=(10,5))
# # #要绘制的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [1,3,2,8,10]
colors=['#66b33f','#99ff99','#ffcc99','#ff9999','#ff4499']
# # #绘制环形图
plt.pie(times,labels=things,
        autopct='%1.1f%%',#显示百分比
        shadow=False,#调整初始化图角度
        startangle=90,#调整初始化图的角度
        colors=colors,#调整饼图配色
        wedgeprops={'width':0.5},#设置圆环宽度
        pctdistance=0.6#设置百分比数字位置
        )
# # # #添加标题
plt.title('每日时间分布',color='green',fontsize=20)
plt.text(0,0,'总计\n100%',ha='center',va='center',fontsize=10)
plt.tight_layout()
plt.show()
```

```python
#绘制散点图
import  matplotlib.pyplot as plt
import random
import numpy as np
from matplotlib import rcParams, colors  # 字体
rcParams['font.sans-serif'] = ['SimHei']
rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
#创建列表、设置大小
plt.figure(figsize=(10,8))
#要绘制的数据
x = []
y = []
for i in range(1000):
    tmp = random.uniform(0.1,10)
    x.append(tmp)
    tmp2=2*tmp + random.gauss(0,2)
    y.append(tmp2)
    print(tmp,tmp2)
#绘制散点图
plt.scatter(x,y,
            color='green',
            alpha=0.8,
            s = 20,
            label='数据'
            )
# 关键：自动拟合真实趋势线
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), color='blue', linewidth=2, label='拟合趋势线')
# #添加标题
plt.title('X与Y变量关系',color='red',fontsize=20)
# #添加坐标轴的标签
plt.xlabel('X自变量',fontsize = 10)
plt.ylabel('Y因变量',fontsize = 10)
# #添加图例
plt.legend(loc='upper left',fontsize = 10)
# #添加网格线
plt.grid(True,alpha=0.1,color='blue',linestyle='--')
# #设置刻度字体大小
plt.xticks(rotation=0,fontsize=10)
plt.yticks(rotation=0,fontsize=10)
# #设置y轴的范围
plt.ylim(0,30)
plt.plot([0,10],[0,20])
plt.show()
```

```python
#分析案例：温度分析
#1.导入库
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
rcParams['font.sans-serif'] = ['SimHei']
#2.导入数据
df=pd.read_csv('data/weather.csv')
df.head()
#绘制气温的气势变化图
df['date']=pd.to_datetime(df['date'])
df = df[df['date'].dt.year==2015]
plt.figure(figsize=(20,10))
plt.plot(df['date'],df['temp_max'],label='最高气温')
plt.plot(df['date'],df['temp_min'],label='最低气温')
plt.title('2015气温趋势变化图')
plt.xlabel('日期')
plt.ylabel('气温')
plt.legend()
```

```python
df['temp_mean'] = (df['temp_max'] + df['temp_min']) / 2
plt.figure(figsize=(20,10))
plt.plot(df['date'],df['temp_mean'],label='平均气温')
plt.title('2015平均气温趋势变化图')
plt.xlabel('日期')
plt.ylabel('气温')
df.head(5)
```
