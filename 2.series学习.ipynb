案例：创建一个包含10名学生数学成绩的Series，成绩范围在50-100之间。
计算平均分、最高分、最低分，并找出高于平均分的学生人数。

```python
#生成随机成绩
np.random.seed(42)
values = np.random.randint(50,101,10)
indexes = []
for i in range(1,11):
    indexes.append('学生'+str(i))
scores = pd.Series(values,indexes)
print(scores)
print('平均分：',scores.mean())
print('最高分：',scores.max())
print('最低分:',scores.min())
#高于平均分的学生人数
mean = scores.mean()
scores[scores>mean].count()
len(scores[scores>mean])
```

给定某城市一周每天的最高温度Series，完成以下任务：
找出温度超过30度的天数
计算平均温度
将温度从高到低排序
找出温度变化最大的两天

```python
temperatures = pd.Series([28,31,209,32,30,27,33],index = ['周一','周二','周三','周四','周五','周六','周日'])
```

```python
#找出温度超过30度的天数
n = temperatures[temperatures>30].count()
print('超过30°的天数：',n)
```

```python
#计算平均温度
print('平均温度：',temperatures.mean())
```

```python
#将温度从低到高排序
t2 = temperatures.sort_values(ascending=False)
print(t2)
```

```python
#找出温度最大的两天
t3 = temperatures.diff().abs()#后一个元素跟前一个差值，计算series数组变化的值
t3.sort_values(ascending=False).index[:2].tolist()
```

给定某股票连续10个交易日的收盘价Series：

```python
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2], index=pd.date_range('2023-01-01', periods=10))
```

```python
hours_sales = pd.Series(np.random.randint(0, 100, 24),
                  index=pd.date_range('2025-01-01', periods=24, freq='h'))
#按天重采样
day_sales = hours_sales.resample('D').sum()
# hours_sales.sum
```

```python
#计算每天营业时间（8:00-22:00）和非营业时间的销售额比例
# hours_sales.between_time('8:00','22:00')#获取一段时间内的series
business_hours_sales = hours_sales[(hours_sales.index.hour>=8) & (hours_sales.index.hour<=22)]
print(business_hours_sales/(day_sales-business_hours_sales))
not_business_hours_sales = hours_sales.drop(business_hours_sales.index)
```
